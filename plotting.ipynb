```julia
globalTemp =[14.4,14.5,14.8,15.2,15.5,15.8]
npirat = [45000,2000,15000,5000,400,17]

#to flip the x axis and to see the data in reverse we can use xflip!()
plot(globalTemp,npirat,legend=false)
scatter!(globalTemp,npirat,legend=false)

xlabel!("Number of Pirates(Approx.)")
ylabel!("Global Temperature (C)")
title!("Influence of pirate population on temperature")
```
